```python
df = spark.read.csv(
    "/Volumes/sample_data/default/sample/sample_data.csv",
    header=True,
    inferSchema=True
)
df.show()
```

```python
df_transformed = df.filter(df["age"]>18)
df_transformed = df_transformed.withColumn("adult",df["age"]>18)
df_transformed.show()
```

```python
df_transformed.write.parquet("/Volumes/sample_data/default/sample/transformed_data/")
```
